```python
#PROJET 3 : CODE REPONSES 1 à 14 ET EXPORTATION DES TABLES EN CSV POUR LES IMPORTER DANS UNE BASE SQL (Q15 à 18)

#IMPORTATION MODULES

import numpy as np
import pandas as pd
import seaborn as sns

pd.set_option('display.max_rows', 400)
pd.set_option('min_rows', 50)
```

```python
#IMPORTATION DONNEES

# indicateurs produits animaux(on rajoute une colonne origine)
animaux = pd.read_csv('donnees_fao/FAOSTAT_animaux.csv')
animaux = animaux[~animaux["Code zone"].isin([96,128,41,214])]
animaux['Origine'] = 'A'
animaux = animaux[animaux["Année"] == 2013]

# indicateurs produits vegetaux (on rajoute une colonne origine)
vegetaux = pd.read_csv('donnees_fao/FAOSTAT_vegetaux.csv')
vegetaux = vegetaux[~vegetaux["Code zone"].isin([96,128,41,214])]
vegetaux['Origine'] = 'V'
vegetaux = vegetaux[vegetaux["Année"] == 2013]

# qté de céréales produites
cereales = pd.read_csv('donnees_fao/FAOSTAT_cereales.csv')
cereales = cereales[~cereales["Code zone"].isin([96,128,41,214])]
cereales = cereales[cereales["Année"] == 2013]

# population mondiale par pays entre 2012 et 2018
population = pd.read_csv('donnees_fao/FAOSTAT_population_2012_2018.csv')

# personnes sous alimentées dans le monde
sous_alim = pd.read_csv('donnees_fao/FAOSTAT_data_sous_alim.csv')
sous_alim = sous_alim[~sous_alim["Code zone"].isin([96,128,41,214])]
#sous_alim = sous_alim[sous_alim["Année"] == 2013]
```

```python
#Q1
# Calcul population totale 2013 et 2018 (en milliers d'habitants)

#on restreint por obtenir les populations pour l'année 2013 et 2018
population_2013 = population[population['Année'] == 2013]
population_2018 = population[population['Année'] == 2018]

#on aggrège les populations de tous les pays
pop_total_2013 = population_2013['Valeur'].sum()
pop_total_2018 = population_2018['Valeur'].sum()

print(f'Population totale en 2013 : {round(pop_total_2013/1000000,3)} milliards d\' habitants')
print(f'Population totale en 2018 : {round(pop_total_2018/1000000,3)} milliards d\' habitants')
```

```python
#Q1
# Calcul corrigé en supprimant les sous-ensembles de la Chine
population_2013_v2 = population_2013[~population_2013["Code zone"].isin([96,128,41,214])]
population_2018_v2 = population_2018[~population_2018["Code zone"].isin([96,128,41,214])]

pop_total_2013_v2 = population_2013_v2['Valeur'].sum()
pop_total_2018_v2 = population_2018_v2['Valeur'].sum()

print(f'Population totale (corrigée) en 2013 : {round(pop_total_2013_v2/1000000,3)} milliards d\' habitants')
print(f'Population totale (corrigée) en 2018 : {round(pop_total_2018_v2/1000000,3)} milliards d\' habitants')
```

```python
#Q2
#Disponibilité intérieure = 
#production + importation - exportations + variation stocks = 
#dispo alimentaire + semences + pertes + aliments pour animaux + traitement + autres utilisations)

# masque sur le df vegetaux pour afficher le blé en France
ble_france = vegetaux[(vegetaux['Zone'] == 'France') & (vegetaux['Produit'] == 'Blé')]
ble_france
```

```python
#Q3
#Question 3 : Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

# on unie les tableaux vegetaux et animaux
bilan_complet = pd.concat([animaux, vegetaux], keys = ['animaux', 'vegetaux'])

# on restreint aux lignes à celles dont le 'Code Élement' est 664 ou 674 ou 645
bilan_dispo_alim = bilan_complet[(bilan_complet['Code Élément'] == 664) | (bilan_complet['Code Élément'] == 674) | (bilan_complet['Code Élément'] == 645)]

#on fait une jointure avec le tableau population
population_2013_v2 = population_2013_v2.rename(columns={'Valeur': 'Population', 'Unité' : 'Unité Population'})#changement nom col
bilan_dispo_alim = pd.merge(bilan_dispo_alim, population_2013_v2[['Code zone', 'Population', 'Unité Population']])

#on simplifie le tableau en ne conservant que les colonnes nécessaires
bilan_dispo_alim = bilan_dispo_alim[['Code zone', 'Zone', 'Code Élément', 'Élément', 'Code Produit',
                                     'Produit', 'Année', 'Unité', 'Valeur', 'Population', 'Unité Population']]

# on rajoute la colonne 'Disponibilité alimentaire totale'
bilan_dispo_alim['Disponibilité alimentaire totale'] = bilan_dispo_alim['Valeur'] 

#ON AJUSTE LES UNITES POUR OBTENIR LES VALEURS TOTALES ANNUELLES EN KG OU KCAL
#on affiche les disponibilités alimentaires (en kg) en annuel
bilan_dispo_alim.loc[bilan_dispo_alim['Code Élément'] == 645, 'Disponibilité alimentaire totale'] = bilan_dispo_alim['Disponibilité alimentaire totale'] * bilan_dispo_alim['Population'] * 1000                                                       
#on affiche les disponibilités de protéines en kg et en annuel
bilan_dispo_alim.loc[bilan_dispo_alim['Code Élément'] == 674, 'Disponibilité alimentaire totale'] = bilan_dispo_alim['Disponibilité alimentaire totale'] * bilan_dispo_alim['Population'] * 1000 * 365 / 1000
#on affiche la disponibilité alimentaire (en Kcal) en annuel
bilan_dispo_alim.loc[bilan_dispo_alim['Code Élément'] == 664, 'Disponibilité alimentaire totale'] = bilan_dispo_alim['Disponibilité alimentaire totale'] * bilan_dispo_alim['Population'] * 1000 * 365

#On trie le tableau par pays
bilan_dispo_alim.sort_values(by = ['Zone','Code Élément'] )

pd.options.display.float_format = '{0:,.2f}'.format #formatage pour afficher les valeurs plus clairement ({:.2E} alt)
bilan_dispo_alim
```

```python
#Q4
#Question 4 : A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire 
#(pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids",
#que vous donnerez en kcal/kg. Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio
#aux données disponibles sur internet, par exemple en cherchant la valeur calorique d'un oeuf.

#on utilise pivot_table pour afficher les dispo alimentaires par produit
bilan_dispo_croise = bilan_dispo_alim.pivot_table(values = 'Disponibilité alimentaire totale', index=['Produit', 'Code Élément', 'Élément'],columns=['Zone'])

#on fait la somme des disponibilités alimentaires et de proteines de tous les pays pour chaque produit
bilan_dispo_croise['Total'] = bilan_dispo_croise.sum(axis = 1)

#on renomme les colonne élement
bilan_dispo_croise = bilan_dispo_croise.rename(index= {'Disponibilité alimentaire (Kcal/personne/jour)': 'Disponibilité alimentaire totale (Kcal)'})
bilan_dispo_croise = bilan_dispo_croise.rename(index= {'Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité totale en quantité (kg)'})
bilan_dispo_croise = bilan_dispo_croise.rename(index= {'Disponibilité de protéines en quantité (g/personne/jour)': 'Disponibilité totale de protéines en quantité (Kg)'})
bilan_dispo_croise
```

```python
#Q4 (suite)

#on créé un nouveau tableau croisé affichant les totaux
bilan_dispo_croise_total = bilan_dispo_croise.pivot_table(values = 'Total', index='Produit',columns='Élément')

#on rajoute la colonne ratio énergie/poids
bilan_dispo_croise_total['Ratio énergie/poids (Kcal/Kg)'] = bilan_dispo_croise_total['Disponibilité alimentaire totale (Kcal)'] \
                                                          / bilan_dispo_croise_total['Disponibilité totale en quantité (kg)']
bilan_dispo_croise_total
```

```python
#Q4 2ème partie : ratio poids proteines/poids total - on reprend la méthodologie précedente

#on réarrange le df
bilan_dispo_ratio = bilan_dispo_alim.pivot_table(values = 'Disponibilité alimentaire totale', index=['Zone', 'Produit'],columns=['Élément'])

#on renomme les colonnes
bilan_dispo_ratio = bilan_dispo_ratio.rename(columns = {'Disponibilité alimentaire (Kcal/personne/jour)': 'Disponibilité alimentaire totale (Kcal)'})
bilan_dispo_ratio = bilan_dispo_ratio.rename(columns = {'Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité totale en quantité (kg)'})
bilan_dispo_ratio = bilan_dispo_ratio.rename(columns = {'Disponibilité de protéines en quantité (g/personne/jour)': 'Disponibilité totale de protéines en quantité (Kg)'})

#on rajoute la colonne ratio de proteines
bilan_dispo_ratio['Ratio de protéines'] = bilan_dispo_ratio['Disponibilité totale de protéines en quantité (Kg)'] / bilan_dispo_ratio['Disponibilité totale en quantité (kg)']

#on affiche le ratio en %
bilan_dispo_ratio.head(200).style.format({'Disponibilité alimentaire totale (Kcal)': "{0:,.0f}",'Disponibilité totale en quantité (kg)': "{0:,.0f}",'Disponibilité totale de protéines en quantité (Kg)': "{0:,.0f}",'Ratio de protéines (%)': "{:.2%}"})
```

```python
#Q5 
#Citez 5 aliments parmi les 20 aliments les plus caloriques, en utilisant le ratio énergie/poids.
#Étonnamment, il arrive que ce ratio soit différent en fonction du pays.
#Il faudra donc réaliser pour chaque aliment une moyenne sur les différents pays.
#Vous créerez donc une nouvelle table grâce à une agrégation.
#Attention à bien retirer les valeurs égales à 0 afin de ne pas fausser le calcul de la moyenne.

#on rajoute le ratio energie/poids par pays et par produit dans le df precedent
bilan_dispo_ratio['Ratio énergie/poids (Kcal/Kg)'] =  bilan_dispo_ratio['Disponibilité alimentaire totale (Kcal)'] \
                                                          / bilan_dispo_ratio['Disponibilité totale en quantité (kg)']

#on créé un nouveau df avec seulement les ratios et sans les nan et les 0 pour ne pas fausser les calculs
df_ratios = bilan_dispo_ratio[['Ratio énergie/poids (Kcal/Kg)', 'Ratio de protéines']]
df_ratios = df_ratios.dropna()
df_ratios =df_ratios[df_ratios['Ratio énergie/poids (Kcal/Kg)'] != 0]

#on aggrege par produit et on fait une moyenne
df_ratios_gb = df_ratios.groupby(['Produit']).mean()
#on trie selon le ratio energie/poids décroissant
df_ratios_gb = df_ratios_gb.sort_values(by = 'Ratio énergie/poids (Kcal/Kg)', ascending=False)
df_ratios_gb.head(20)
```

```python
#Q5 partie 2
#Citez 5 aliments parmi les 20 aliments les plus riches en protéines.

#on trie le df par ratio de protéines décroissant
df_ratios_gb_2 = df_ratios_gb.sort_values(by = 'Ratio de protéines', ascending=False)
df_ratios_gb_2.head(20)
```

```python
#Q6
#Calculez, pour les produits végétaux uniquement, la disponibilité intérieure mondiale exprimée en kcal.


#on crée un df comprenant les demandes interieures(5301) pour les vegetaux en 2013
dispo_int_vegetaux = bilan_complet[(bilan_complet['Code Élément'] == 5301) & (bilan_complet['Origine'] == 'V') & (bilan_complet['Année'] == 2013)]

#on utilise pivot_table pour afficher les dispo interieures par produit et pouvoir calcluler le total des pays en Kg
dispo_int_vegetaux_2 = dispo_int_vegetaux.pivot_table(values = 'Valeur', index=['Produit'],columns=['Zone'])
dispo_int_vegetaux_2['Disponibilité intérieure mondiale (Kg)'] = dispo_int_vegetaux_2.sum(axis = 1) * 1000000

#on créé un df dispo interieure mondiale
dispo_int_vegetaux_tot = dispo_int_vegetaux_2[['Disponibilité intérieure mondiale (Kg)']]

#on fait une jointure interne avec le tableau des ratios
dispo_int_ratios = df_ratios_gb.join(dispo_int_vegetaux_tot, how = 'inner')
dispo_int_ratios = dispo_int_ratios.sort_values(by = 'Produit')

#on ajoute la colonne disponibilité intérieure mondiale en Kcal
dispo_int_ratios['Disponibilité intérieure mondiale (Kcal)'] = dispo_int_ratios['Disponibilité intérieure mondiale (Kg)'] * dispo_int_ratios['Ratio énergie/poids (Kcal/Kg)']
dispo_int_ratios
```

```python
#Q7
#Combien d'humains pourraient être nourris si toute la disponibilité intérieure mondiale de produits végétaux était utilisée
#pour de la nourriture ? Donnez les résultats en termes de calories, puis de protéines,
#et exprimez ensuite ces 2 résultats en pourcentage de la population mondiale.

#on calcule le total de la disponibilité intérieure en calories en faisant la somme pour chaque produit
dispo_int_totale = dispo_int_ratios['Disponibilité intérieure mondiale (Kcal)'].sum()
print(f'Disponibilité intérieure anuelle pour les vegetaux : {dispo_int_totale} Kcal')
print(f'Disponibilité intérieure journalière pour les vegetaux : {dispo_int_totale/365} Kcal \n')

#on rajoute la colonne disponibilité intérieure mondiale en Kg de protéines
dispo_int_ratios['Disponibilité intérieure mondiale en protéines (Kg)'] = dispo_int_ratios['Disponibilité intérieure mondiale (Kg)'] * dispo_int_ratios['Ratio de protéines']
dispo_int_totale_prot = dispo_int_ratios['Disponibilité intérieure mondiale en protéines (Kg)'].sum()

# en supposant une moyenne de 1800 Kcal ou 50g de proteines nécessaire à un humain pour survivre
cal_min = 1800 #en Kcal
prot_min = 50 /1000 #en Kg
humains_nourris_kcal = (dispo_int_totale/365) /cal_min
humains_nourris_prot = (dispo_int_totale_prot/365) /prot_min

print('En se basant sur 1800 Kcal par jour :')
print(f'{round(humains_nourris_kcal/1000000000,3)} milliards d\'humains, soit {round((humains_nourris_kcal/(pop_total_2013_v2*1000))*100,1)} % de la population actuelle (2013) pourraient être nourris si la disponibilité intérieure mondiale de végetaux était utilisée pour de la nourriture.\n')
print('En se basant sur 50g de protéines par jour :')
print(f'{round(humains_nourris_prot/1000000000,3)} milliards d\'humains, soit {round((humains_nourris_prot/(pop_total_2013_v2*1000))*100,1)} % de la population actuelle (2013) pourraient être nourris si la disponibilité intérieure mondiale de végetaux était utilisée pour de la nourriture.') 
```

```python
#Q8
#Combien d'humains pourraient être nourris si toute la disponibilité alimentaire en produits végétaux,
#la nourriture végétale destinée aux animaux et les pertes de produits végétaux étaient utilisés pour de la nourriture ? 
#Donnez les résultats en termes de calories, puis de protéines, et exprimez ensuite ces 2 résultats en % de la pop mondiale.

#on reprend la méthodologie de la Q6 et Q7 en remplacant disponibilité interieure par 
#l'aggregat (disponibilité alimentaire(645) + aliments pour animaux(5521) + pertes(5123) qu'on appelle dispo alimentaire étendue

#on crée un df comprenant les differents aggrégats demandés pour les vegetaux en 2013
dispo_alim_etendue = bilan_complet[(bilan_complet['Code Élément'].isin([645, 5521, 5123])) & (bilan_complet['Origine'] == 'V') & (bilan_complet['Année'] == 2013)]

#on fait une jointure avec le tableau population
dispo_alim_etendue = pd.merge(dispo_alim_etendue, population_2013_v2[['Code zone', 'Population', 'Unité Population']])

#on ajuste les unités
dispo_alim_etendue.loc[dispo_alim_etendue['Code Élément'] == 645, 'Valeur'] = dispo_alim_etendue['Valeur'] * dispo_alim_etendue['Population'] * 1000              
dispo_alim_etendue.loc[dispo_alim_etendue['Code Élément'].isin([5521, 5123]), 'Valeur'] = dispo_alim_etendue['Valeur'] * 1000000              

#on utilise pivot_table pour afficher les dispo interieures par produit et pouvoir calcluler le total des pays en Kg
dispo_alim_etendue_2 = dispo_alim_etendue.pivot_table(values = 'Valeur', index=['Produit', 'Élément'],columns=['Zone'])
dispo_alim_etendue_2['Disponibilité totale en Kg'] = dispo_alim_etendue_2.sum(axis = 1)

#on réarrange le df et on ajoute une colonne Disponibilité alimentaire étendue
dispo_alim_etendue_3 = dispo_alim_etendue_2.pivot_table(values = 'Disponibilité totale en Kg', index=['Produit'],columns=['Élément'])
dispo_alim_etendue_3['Disponibilité alimentaire étendue (Kg)'] = dispo_alim_etendue_3.sum(axis = 1)

#on fait une jointure interne avec le tableau des ratios
dispo_alim_ratios = df_ratios_gb.join(dispo_alim_etendue_3, how = 'inner')
dispo_alim_ratios = dispo_alim_ratios.sort_values(by = 'Produit')

#on ajoute la colonne disponibilité alimentaire étendue (Kcal) mondiale en Kcal
dispo_alim_ratios['Disponibilité alimentaire étendue (Kcal)'] = dispo_alim_ratios['Disponibilité alimentaire étendue (Kg)'] * dispo_int_ratios['Ratio énergie/poids (Kcal/Kg)']

dispo_alim_ratios = dispo_alim_ratios.rename(columns = {'Disponibilité alimentaire en quantité (kg/personne/an)': 'Disponibilité alimentaire en quantité (kg)'}) # on renomme la colonne pour préciser l'unité
dispo_alim_ratios
```

```python
#Q8 suite

#on calcule le total de la disponibilité alimentaire étendue en calories en faisant la somme pour chaque produit
dispo_alim_totale = dispo_alim_ratios['Disponibilité alimentaire étendue (Kcal)'].sum()

#on rajoute la colonne disponibilité alimentaire étendue mondiale en Kg de protéines
dispo_alim_ratios['Disponibilité alimentaire étendue mondiale en protéines (Kg)'] = dispo_alim_ratios['Disponibilité alimentaire étendue (Kg)'] * dispo_alim_ratios['Ratio de protéines']
dispo_alim_totale_prot = dispo_alim_ratios['Disponibilité alimentaire étendue mondiale en protéines (Kg)'].sum()

# en supposant une moyenne de 1800 Kcal ou 50g de proteines nécessaire à un humain pour survivre
humains_nourris_kcal_2 = (dispo_alim_totale/365) /cal_min
humains_nourris_prot_2 = (dispo_alim_totale_prot/365) /prot_min

print('En se basant sur 1800 Kcal par jour :')
print(f'{round(humains_nourris_kcal_2/1000000000,3)} milliards d\'humains, soit {round((humains_nourris_kcal_2/(pop_total_2013_v2*1000))*100,1)} % de la population actuelle (2013) pourraient être nourris si la disponibilité alimentaire en végetaux, l\'alimentation végétale pour les animaux et les pertes étaient utilisées pour de la nourriture.\n')
print('En se basant sur 50g de protéines par jour :')
print(f'{round(humains_nourris_prot_2/1000000000,3)} milliards d\'humains, soit {round((humains_nourris_prot_2/(pop_total_2013_v2*1000))*100,1)} % de la population actuelle (2013) pourraient être nourris si la disponibilité alimentaire en végetaux, l\'alimentation végétale pour les animaux et les pertes étaient utilisées pour de la nourriture.') 
```

```python
#Q12 (suite)

#on restreint le bilan complet aux produits trouvés précedemment et aux importations (5611)
pays_sous_alim_export_15 = pays_sous_alim_export_15.reset_index()
bilan_import = bilan_complet[(bilan_complet['Produit'].isin(pays_sous_alim_export_15['Produit'])) \
                             & (bilan_complet['Code Élément'] == 5611)].copy()

#on selectionne les 200 plus grandes importations
bilan_import.sort_values(by = 'Valeur', ascending = False, inplace = True) 
bilan_import_200 = bilan_import.head(200).copy()
bilan_import_200

#on groupe les importations par produit
bilan_import_200_gb = bilan_import_200.groupby('Produit')[['Valeur']].sum()
bilan_import_200_gb = bilan_import_200_gb.rename(columns = {'Valeur': 'Importations (milliers de tonnes)'})
bilan_import_200_gb
```

```python
#Q12 (suite)

#on crée un df avec les quantités autres utilisations, dispo interieure, aliments pour animaux et nourriture pour les 200 couples zone/produit
bilan_200 = bilan_complet.merge(bilan_import_200[['Zone', 'Produit']], on = ['Zone', 'Produit'] )
bilan_200 = bilan_200[bilan_200['Code Élément'].isin([5154, 5301, 5521, 5142])]

#on calcule les totaux pour chaque élement puis on l'ajoute au df bilan_import_200_gb

bilan_200_a = bilan_200[bilan_200['Code Élément'] == 5154] #Autres utilisations
bilan_import_200_a_gb = bilan_200_a.groupby('Produit')[['Valeur']].sum()
bilan_import_200_gb['Autres utilisations (milliers de tonnes)'] = bilan_import_200_a_gb['Valeur']

bilan_200_b = bilan_200[bilan_200['Code Élément'] == 5301] #dispo interieure
bilan_import_200_b_gb = bilan_200_b.groupby('Produit')[['Valeur']].sum()
bilan_import_200_gb['Disponibilité intérieure (milliers de tonnes)'] = bilan_import_200_b_gb['Valeur']

bilan_200_c = bilan_200[bilan_200['Code Élément'] == 5521] #aliments pour animaux
bilan_import_200_c_gb = bilan_200_c.groupby('Produit')[['Valeur']].sum()
bilan_import_200_gb['Aliments pour animaux (milliers de tonnes)'] = bilan_import_200_c_gb['Valeur']

bilan_200_d = bilan_200[bilan_200['Code Élément'] == 5142] #nourriture
bilan_import_200_d_gb = bilan_200_d.groupby('Produit')[['Valeur']].sum()
bilan_import_200_gb['Nourriture (milliers de tonnes)'] = bilan_import_200_d_gb['Valeur']
bilan_import_200_gb.fillna(0, inplace = True)

#on calcule les ratios
bilan_import_200_gb['Ratio Autres utilisations / Disponibilité intérieure (%)'] = (bilan_import_200_gb['Autres utilisations (milliers de tonnes)'] / bilan_import_200_gb['Disponibilité intérieure (milliers de tonnes)']) *100
bilan_import_200_gb['Ratio Alimentation animale / Alimentation totale (%)'] = (bilan_import_200_gb['Aliments pour animaux (milliers de tonnes)'] / (bilan_import_200_gb['Aliments pour animaux (milliers de tonnes)'] + bilan_import_200_gb['Nourriture (milliers de tonnes)'])) *100
bilan_import_200_gb
```
